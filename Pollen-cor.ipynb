```python
L=['Frêne','Olivier','Noisetier de Byzance','Bouleau','Charme','Cyprès','Mûrier','Platane','Aulne']
```

```python
L
```

```python
T1=T[T["DOMANIALITE"]=="Alignement"]
```

```python
T1.head()
```

```python
def proportion(colonne,liste):
    n=0
    for arbre in colonne:
        if arbre in liste:
            n=n+1
    return n/colonne.count()
```

```python
proportion(T1['LIBELLEFRANCAIS'],L)
```

La proportion d'arbres à fort potentiel allergisant est d'environ 35,6%

```python
def coordonnées(s):
    return float(s.split(',')[0]),float(s.split(',')[1])
```

```python
coordonnées('45.3,2.6')
```

```python
type(coordonnées('45.3,2.6'))
```

```python
Loc=[]
for k in T1.index.values:
    if T1.loc[k,'LIBELLEFRANCAIS'] in L:
        Loc.append(coordonnées(T1.loc[k,'geo_point_2d']))
```

```python
import folium
from folium import plugins
```
